# Calorically dense foods

Suppose you are given the following dataframe containing food, weight, and calories. You'll notice the foods have varying weights associated with them:

image.png

Using Python (Pandas), can you sort the dataframe in descending order with the most calorically dense food (normalized for weight) at the top?

```python
## Import modules
import pandas as pd
```

```python
## Build dataset based on image
food = ['bacon','strawberries','banana','spinach','chicken breast','peanuts']
grams = [50,200,100,200,50,100]
calories = [271,64,89,46,80,567]

df = pd.DataFrame({
    'food': food,
    'grams': grams,
    'calories': calories
})

df
```

```python
## Create column for calorie density
df['density'] = df.calories * 1.0 / df.grams

df
```

```python
## Sort it descending
df.sort_values(by=['density'], ascending = False)
```
